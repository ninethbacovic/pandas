## 2- fillna - Replace the NaN values for a meaningfull value

```python
# In this case the NaN is replaced by zero
new_df = df.fillna(0)
new_df
```

Sometimes having zero in all the cells is not he best choice,  For example, in the event, you can't have a type of weather as zero.

```python
# Using a dictionary
new_df = df.fillna({
        'temperature' : 0,
        'windspeed' : 0,
        'event' : 'no event'
    })
new_df
```

```python
# Repeat the value before and carry it forward
new_df = df.fillna(method = 'ffill')
new_df
```

```python
# Repeat the value after and carry it forward
new_df = df.fillna(method = 'bfill', limit = 2) #limit the copies
new_df
```

## 3 - Interpolate values

```python
#Interpolate the values (its a linear interpolation --> x-32 = 28-x, x=30)
new_df = df.interpolate()
new_df
```

```python
#Interpolate the values with the method time, because the dates are different 
#if you notice january 4th is nearst 5th than 1st... so its value maybe must be
#closer to 5th day, even in temperature or in the other columns 
new_df = df.interpolate(method = 'time')
new_df
```

## 4 - Dropna - takes away the rows with NaN values


```python
#Drop the rows with at least one NaN value
new_df = df.dropna() 
new_df
```

```python
#just drop the row with all NaN values
new_df = df.dropna(how = 'all') 
new_df
```

```python
# another way to drop the row with all NaN values
new_df = df.dropna(thresh = 1) 
new_df
```

```python
# another way to drop the row with at least 2 VALID values
new_df = df.dropna(thresh = 2) 
new_df
```

## 5 - Insert missing dates

```python
#Using a data range 
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
