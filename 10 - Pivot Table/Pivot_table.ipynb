## 1- Pivot
Pivot allows you to transform or reshape data

```python
import pandas as pd
df = pd.read_csv('weather.csv')
df
```

```python
df.pivot(index = 'date', columns = 'city')
```

```python
df.pivot(index = 'date', columns = 'city', values = 'humidity')
```

```python
df.pivot(index = 'humidity', columns = 'city')
```

## 2- Pivot Table
Pivot table is used to summarize and aggregate data inside dataframe

```python
df = pd.read_csv('weather2.csv')
df
```

For New York, in the first date, I have two temperatures and for the other cities too.

```python
df.pivot_table(index = 'city', columns = 'date')
```

It shows the average of temperatures in each city and date, because the mean is a default

```python
#I can change my agregate function to a sum, for example.
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'sum')
```

```python
#I can change my agregate function to a count, for example.
df.pivot_table(index = 'city', columns = 'date', aggfunc = 'count')
```

```python
df.pivot_table(index = 'city', columns = 'date', margins = True)
```

This is showing the average of all
